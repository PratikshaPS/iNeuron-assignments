```python
n=int(input("Enter number here: "))
for i in range(n):
    for j in range(i):
        print("*", end=" ")
    print(end="\n")
for k in range(n):
    for l in range(n-k):
        print("*", end=" ")
    print(end="\n")
```

#### 2. Write a Python program to reverse a word after accepting the input from the user.
Sample Output:
Input word: ineuron
Output: norueni

```python
word=input("Input word : ")
print("Output : " + word[::-1])
```

```python
word=input("Input word : ")
output=""
for i in range(len(word)):
    output = output + word[len(word)-(i+1)]
    
print("Output : " + output)
```
